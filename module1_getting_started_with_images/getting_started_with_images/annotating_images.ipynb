# Annotating Images

In this notebook we will demonstrate how to annotate images using OpenCV. We will learn how to perform the following annotations to images.

+ Draw lines
+ Draw circles
+ Draw rectangles
+ Add text

These are useful when you want to annotate your results for presentations or show a demo of your application. Annotations can also be useful during development and debugging. Videos can also be annotated as we will see in future modules.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["image.cmap"] = "gray"
```

### Read Grayscale Image as "Color"

```python
# Read the image.
image = cv2.imread("../module01/Apollo-8-Launch.jpg", cv2.IMREAD_COLOR)

# Display the original image.
plt.figure(figsize=(7, 7))
plt.imshow(image)
print(image.shape)
```

```python
# Define colors for annotations.
yellow = (0, 255, 255)
red = (0, 0, 255)
magenta = (255, 0, 255)
green = (0, 255, 0)
```

### Drawing a Line

The **line()** function in OpenCV is used to draw lines on images.

This fuction has four required arguments:

1. img: The image we want to annotate
2. pt1: Starting point for the line in pixel coordinates on the image
3. pt2: Ending point for the line in pixel coordinates on the image
4. color: The color for the line

All these annotations functions require the specification of coordinates in a XY format, unlike the convention for indexing NumPy arrays with the Y coordinate first and then the X coordinate, these particular functions require X first and then Y.

```python
# Make a copy of the original image.
image_line = image.copy()

# Draw a yellow line (using: cv2.Line_8)
image_line = cv2.line(image_line, (0, 455), (450, 465), yellow, thickness=3, lineType=cv2.LINE_8)

# Display the annotated image.
plt.figure(figsize=(10, 10))
plt.imshow(image_line[:, :, ::-1]);
```

```python
# Make a copy of the original image.
image_line_aa = image.copy()

# Draw a yellow line (using: cv2.LINE_AA)
image_line_aa = cv2.line(image_line_aa, (0, 455), (450, 465), yellow, thickness=3, lineType=cv2.LINE_AA)

# Display the annotated image.
plt.figure(figsize=(10, 10))
plt.imshow(image_line_aa[:, :, ::-1]);
```

### Drawing a Circle

The **circle()** fuction in OpenCV is used to draw circles on images.

The required arguments are:

1. img: The image we want annotate
2. center: The coordinates for the center of the circle
3. radius: The radius of the circle in pixels
4. color: The color for the circle

```python
# Make a copy of the original iamge.
image_circle = image.copy()

# Draw a red circle.
image_circle = cv2.circle(image_circle, (195, 55), 20, red, thickness=2, lineType=cv2.LINE_AA)

# Display the annotated image.
plt.figure(figsize=(8, 8))
plt.imshow(image_circle[:, :, ::-1]);
```

### Drawing a rectangle

The **rectangle()** function in OpenCV is used to draw rectangles on images.

The required arguments are:

1. img: The image we want to annotate
2. pt1: Upper left portion of the rectangle
3. pt2: Lower right portion of the rectangle

```python
# Make a copy of the original image.
image_rectangle = image.copy()

# Draw a magenta rectangle.
image_rectangle = cv2.rectangle(image_rectangle, (300, 150), (480, 420), magenta, thickness=2, lineType=cv2.LINE_8)

# Display the annotated image.
plt.figure(figsize=(8, 8))
plt.imshow(image_rectangle[:, :, ::-1]);
```

### Adding Text

The **putText()** function in OpenCV is used to add text to images.

The require arguments are:

1. img: The image we want put the text
2. text: The text string that we want put on the image
3. org: Coordinates for the origin of the bottom left hand corner of the text string

```python
# Make a copy of the original image.
image_text = image.copy()

# Add text to teh image.
text = "Apollo 8 Saturn V Launch, 21 Dec 1968"
font_face = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.8
font_color = green
font_thickness = 1

image_text = cv2.putText(image_text, text, (45, 30), font_face, font_scale, font_color, font_thickness, cv2.LINE_AA)

# Display the annotated image.
plt.figure(figsize=(10, 10))
plt.imshow(image_text[:, :, ::-1]);
```
