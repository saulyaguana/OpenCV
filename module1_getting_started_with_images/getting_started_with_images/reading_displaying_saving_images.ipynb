### Reading images with OpenCV

The function to read images is **imread()**

The arguments are:

+ filename: The path where the image is
+ flags: Allow us to specify how actually read in the image (for example grayscale or color)

We tipically represent images as eight bit unsigned integers and therefore the value range on images is from 0 to 255, where 0 represents pure black and 255 represents pure white.

```python
# Read image as grayscale.
bw_img = cv2.imread("../module01/img_bw_18x18.png", cv2.IMREAD_GRAYSCALE)

# Print the image data (pixel values) of a 2D NumPy array.
# Each pixel value is 8-bits in the range [0, 255].
print(bw_img)
```

```python
# Print the size of image.
print(f"Image size is: {bw_img.shape}")

# Print data-type of image.
print(f"Data type of image is: {bw_img.dtype}")
```

### Display image data as an "image" using Matplotlib

The reason for this color is that matplotlib uses color maps to represent the color information and numpy arrays. In this case the color map was set to something other than grayscale.

```python
# Display the image.
plt.imshow(bw_img);
```

```python
# Set color map to gray scale for proper rendering.
plt.imshow(bw_img, cmap="gray");
```

To save data in the file system we use the opencv function **imwrite()**

Takes two required arguments:

+ filename: The file name you want to save the image
+ img: The numpy array that contains the image data

```python
image = cv2.imread("../module01/Apollo-8-Launch.jpg")
plt.figure(figsize=(10, 10))
plt.imshow(image, cmap="gray")
plt.title("Apollo-8-Launch.jpg", fontsize=16);
```

```python
image.shape
```

### Save the image as PNG file

```python
# Save the image to the file system.
cv2.imwrite("Apollo-8-Launch-img.png", image)
```

### Read the saved image as a PNG file

```python
image = cv2.imread("./Apollo-8-Launch-img.png", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(10, 10))
plt.imshow(image, cmap="gray")
plt.title("Apollo-8-Launch-img.png");
```
