# Human-computer interaction (HCI)


**Human-computer interaction (HCI)** focuses on the interfaces between people and computers. Initially concerned with personal computers, HCI now has evolved into a multidisciplinary field, combining practices from a number of fields including:

+ Computer Science
+ Psychology
+ Design
+ Sociology
+ Ergonomics

and many more.



---

What is Human-Computer Interaction?

In essence it is the interface between people and computers, the interface is not a GUI where you have buttons and the like, but rather the physical interaction between the human and the computer's devices, this could be in a form of a keyboard or a motion tracking system for instance.

# PyAutoGUI

*PyAutoGUI* is python library that can be used to control the mouse and keyboard and it let us automate the task of interacting with computer the way that a human normally would.

```python
import pyautogui
import time
import math
```

# Mouse Control Functions

---

### Mouse Size and Mouse Position:

Locations on your screen are refered to by X and Y Cartesian coordinates.

The screen resolution size is returned by the **size()** function as a tuple of two integers.

```python
# Wait 2 seconds, to give you time to switch to the drawing application.
time.sleep(2.0)

distance = 200
while distance > 0:
    pyautogui.drag(distance, 0, button="left", duration=0.5)  # move right
    distance -= 50
    pyautogui.drag(0, distance, button="left", duration=0.5)  # move down
    pyautogui.drag(-distance, 0, button="left", duration=0.5)  # move left
    distance -= 50
    pyautogui.drag(0, -distance, button="left", duration=0.5)  # move up
```

```python
pyautogui.moveTo(100, 100, duration=3, tween=pyautogui.easeOutQuad)
```

```python
pyautogui.dragTo(100, 100, duration=3, button="left")
```

```python
pyautogui.click(300, 200, clicks=2, interval=1, button="left", duration=4, tween=pyautogui.easeInOutQuad)
```

```python
time.sleep(2)


x_c, y_c = 600, 400
r = 100


pyautogui.moveTo(x_c + r, y_c)
for angle in range(0, 361):
    x = x_c + r * math.cos(math.radians(angle))
    y = y_c + r * math.sin(math.radians(angle))
    pyautogui.dragTo(x, y, duration=0.01, button='left')
```

```python
time.sleep(2)

x_c, y_c = 800, 500
a, b = 150, 75

pyautogui.moveTo(x_c + a, y_c)
for angle in range(0, 361):
    x = x_c + a * math.cos(math.radians(angle))
    y = y_c + b * math.sin(math.radians(angle))
    pyautogui.dragTo(x, y, duration=0.005, button='left')
```

#  Parmetros de funciones PyAutoGUI

| Parmetro | Funciones donde aparece | Tipo | Descripcin / Significado |
|------------|------------------------|-------|----------------------------|
| **x** | `moveTo`, `dragTo`, `click` | `int` / `float` | Coordenada horizontal del cursor en pxeles. Define la posicin destino. |
| **y** | `moveTo`, `dragTo`, `click` | `int` / `float` | Coordenada vertical del cursor. |
| **duration** | `moveTo`, `dragTo`, `click` | `float` | Tiempo en segundos que tarda el movimiento hasta el destino. Si es 0, se mueve instantneamente. |
| **tween** | `moveTo`, `dragTo`, `click` | `function` | Funcin de suavizado del movimiento (`pyautogui.linear`, `easeInOutQuad`, etc.). |
| **button** | `dragTo`, `click` | `str` | Botn del mouse a usar: `'left'`, `'right'` o `'middle'`. |
| **clicks** | `click` | `int` | Nmero de clics a realizar (por defecto 1). |
| **interval** | `click`, `press`, `write`, `hotkey` | `float` | Tiempo entre clics, pulsaciones o caracteres. Simula velocidad humana. |
| **presses** | `press` | `int` | Nmero de veces que se presionar una tecla. |
| **message** | `write` | `str` / `list` | Texto a escribir (cadena o lista de teclas). |
| **key** | `press`, `keyDown`, `keyUp` | `str` | Nombre de la tecla a presionar o soltar (`'enter'`, `'a'`, `'ctrl'`, etc.). |
| **\*keys** | `hotkey` | `*args` (lista de strings) | Varias teclas a presionar en combinacin (por ejemplo: `('ctrl', 'shift', 'n')`). |
| **logScreenshot** | `moveTo`, `dragTo`, `click` | `bool` | Si est en `True`, guarda capturas de depuracin. Normalmente no se usa. |
| **_pause** | Todas las funciones principales | `bool` | Si est en `True`, aplica una pausa automtica (definida en `pyautogui.PAUSE`) despus de ejecutar la funcin. |

---

##  Parmetros exclusivos

| Parmetro | Solo pertenece a | Descripcin |
|------------|------------------|--------------|
| **clicks** | `click()` | Define el nmero de clics en una misma accin (simple, doble, triple clic, etc.). |
| **message** | `write()` | Texto o lista de teclas a escribir. |
| **key** | `keyDown()`, `keyUp()` | La tecla a mantener o liberar. |
| **\*keys** | `hotkey()` | Permite pasar varias teclas a la vez (`ctrl`, `alt`, `del`). |
| **button** | `dragTo()` y `click()` | Determina qu botn del mouse se mantiene presionado o hace clic. |
| **tween** | `moveTo()`, `dragTo()`, `click()` | Define la interpolacin de movimiento (animacin del mouse). |

---

##  Parmetros comunes (comparten varias funciones)

| Parmetro | Significado comn | Usado en |
|------------|-------------------|-----------|
| **duration** | Tiempo que tarda la accin (movimiento o animacin). | `moveTo`, `dragTo`, `click` |
| **interval** | Tiempo entre acciones repetidas (clics, teclas, letras). | `click`, `press`, `write`, `hotkey` |
| **_pause** | Pausa global entre llamadas a funciones PyAutoGUI. | Todas |
| **x**, **y** | Coordenadas absolutas. | `moveTo`, `dragTo`, `click` |
